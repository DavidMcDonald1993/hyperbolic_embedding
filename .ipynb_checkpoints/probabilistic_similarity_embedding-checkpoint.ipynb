```python
from scipy.sparse import identity, diags

N = nx.number_of_nodes(G)

# A = np.array(nx.adjacency_matrix(G).todense())
# DZ = np.array(AZ.sum(axis=0)).flatten()

W = csr_matrix(AZ / AZ.sum(axis=0))
# lazy random walk
W = (identity(n=N) + W) / 2


def matrix_multiply(M, n):
    if n == 0:
        return identity(M.shape[0])
    if n % 2 == 0:
        m = matrix_multiply(M, n/2).dot(matrix_multiply(M, n/2))
    else: 
        m = M.dot(matrix_multiply(M, n-1))
    return m
```

```python
T = 0.5
degrees = np.array(nx.degree(G).values())
m = degrees.mean() / 2
```

```python
degree_sort = degrees.argsort()[::-1]
```

```python
ranked_nodes = np.array([np.where(degree_sort==x)[0][0] for x in range(N)])
```

```python
import powerlaw

result = powerlaw.Fit(degrees)

gamma = result.alpha

beta = 1. / (gamma - 1)

R = 2 * beta * np.log(range(1, N + 1)) + 2 * (1 - beta) * np.log(N) 
R = R[ranked_nodes]
```

```python
from keras.models import Model
from keras.layers import Input, Dense, Dropout, Lambda, BatchNormalization
from keras.regularizers import l2

import keras.backend as K
```

```python
def compute_disk_radius(args):
    
    r_t, i = args
    
    r_t = K.abs(r_t)
    
    return r_t - 2 * K.log((2 * T * (1 - K.exp(- (1 - beta) * K.log(i)))) / (m * (1 - beta) * K.sin(np.pi * T)))

def cosh(x):
    return (K.exp(x) + K.exp(-x)) / 2

def sinh(x):
    return (K.exp(x) - K.exp(-x)) / 2

def arccosh(x):
    return K.log(x + K.sqrt(x + 1) * K.sqrt(x - 1))

def hyperbolic_distance(args):
    
    r1, r2, theta1, theta2  = args
    
    # compute hyperbolic distance 
    pi = K.constant(np.pi)
    delta = pi - K.abs(pi - K.abs(theta1 - theta2))
    d = cosh(r1) * cosh(r2) - sinh(r1) * sinh(r2) * K.cos(delta)
    d = K.maximum(1.0, d)
    d = arccosh(d)
    
    return d

def compute_probability(args):
    disk_R, d = args
    return K.sigmoid((disk_R - d) / (2 * T))
#     return K.sigmoid(-d)
```

```python
# contruct model to determine angular coordinates from one hot representation of node
dropout = 0.1
num_hidden = [64]
reg = l2(1e-3)

x = Input(shape=(N,))

y = Dense(num_hidden[0], activation="relu", kernel_regularizer=reg)(x)
y = BatchNormalization()(y)
y = Dropout(dropout)(y)


for h in num_hidden[1:]:
    y = Dense(h, activation="relu", kernel_regularizer=reg)(y)
    y = BatchNormalization()(y)
    y = Dropout(dropout)(y)
    
theta = Dense(1, activation="linear", kernel_regularizer=reg)(y)

angular_embedder = Model(x, theta)

# y = Dense(2, activation="linear")(y)
# r = Lambda(lambda x : x[:,:1])(y)
# theta = Lambda(lambda x : x[:,1:])(y)

# angular_embedder = Model(x, [r, theta])
```

```python
# embedder of two nodes
r1 = Input(shape=(1,))
r2 = Input(shape=(1,))

x1 = Input(shape=(N,))
x2 = Input(shape=(N,))

i = Input(shape=(1,))

theta1 = angular_embedder(x1)
theta2 = angular_embedder(x2)

disk_radii = Lambda(compute_disk_radius, output_shape=(1,))([r1, i])

hyperbolic_distances = Lambda(hyperbolic_distance, output_shape=(1,))([r1, r2, theta1, theta2])
probabilities = Lambda(compute_probability, output_shape=(1,))([disk_radii, hyperbolic_distances])

probability_model = Model([r1, r2, x1, x2, i], probabilities)
probability_model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
from keras_tqdm import TQDMNotebookCallback
tqdm = TQDMNotebookCallback(leave_outer=True, leave_inner=False)

from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(min_delta=0, monitor="loss", patience=100)
```

```python
from scipy.sparse import find
```

```python
def generator(batch_size=100, num_pos=1, num_neg=5, eps=1e-8):
    
    while True:
        
        nodes = np.random.choice(N, size=batch_size)
        
        # positive samples
        pos = np.array([np.random.choice(N, p=targets[:,n].toarray().flatten(), size=num_pos) for n in nodes])
        
        # negative samples 
        neg = np.array([np.random.choice(find(targets[:,n] < eps)[0], size=num_neg) for n in nodes])
        
        nodes1 = np.repeat(nodes, num_pos + num_neg)
        
        nodes2 = np.array([np.append(pos[i], neg[i]) for i in range(batch_size)]).flatten()

        targs = np.array([np.append(np.ones(num_pos), np.zeros(num_neg)) for i in range(batch_size)]).flatten()
        
#         yield [I[nodes1], I[nodes2], ranked_nodes[nodes1]+1], targs       
        yield [R[nodes1], R[nodes2], I[nodes1], I[nodes2], ranked_nodes[nodes1]+1], targs

def random_pos_neg_generator(batch_size=100, neg_bias=0.8, eps=1e-8):
    
    while True:
        
        nodes = np.random.choice(N, size=batch_size, replace=True)
     
        targs = np.random.choice(2, p=[neg_bias, 1-neg_bias], size=batch_size)
        
        nodes2 = np.array([np.random.choice(N, p=targets[:,nodes[i]].toarray().flatten()) if targs[i] == 1 else
                          np.random.choice(find(targets[:,nodes[i]] < eps)[0]) for i in range(batch_size)])
        
        yield [R[nodes], R[nodes2], I[nodes], I[nodes2]], targs
#         yield [I[nodes], I[nodes2]], targs
```

```python
def hyperbolic_distance_numpy(r1, theta1, r2, theta2):
    
    delta = np.pi - np.abs(np.pi - np.abs(theta1 - theta2))
    d = np.cosh(r1) * np.cosh(r2) - np.sinh(r1) * np.sinh(r2) * np.cos(delta)
    d = np.maximum(1.0, d)
    d = np.arccosh(d)
    
    if np.array_equal(r1, r2) and np.array_equal(theta1, theta2):
        d = 0
    
    return d

D = np.array([[hyperbolic_distance_numpy(r1, theta1, r2, theta2) for r2, theta2 in zip(R.flatten(), thetas.flatten())]
             for r1, theta1 in zip(R.flatten(), thetas.flatten())])
```
